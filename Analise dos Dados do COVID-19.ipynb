## Vamos Análisar o número de mortes

Nessa etapa vamos análisar o número de mortes no período analisado tanto o número total de mortes como também de casos confirmados e os paises afetados

```python
# Total de Mortes
total_de_mortes = dados_covid['Deaths'].sum()

# Exibindo a série
total_de_mortes
```

```python
# Total de mortes por país
mortes_por_paises = dados_covid.groupby('Country')['Deaths'].sum()

# Exibindo a série
mortes_por_paises.sort_values(ascending=False)

# Exportando o dataframe
mortes_por_paises.to_csv('total-de-mortes-por-paises.csv')
```

Com base nos Dados acima vimos que no perído de 187 dias o COVID-19 matou 43.384.903 de pessoas dos 187 paises analisados do dataframe

```python
# Exibindo o país com o número de mortos
pais_com_maior_numero_de_mortes = dados_covid.sort_values('Deaths', ascending=False)

# Exibindo a série
pais_com_maior_numero_de_mortes

# Exportando os dados
pais_com_maior_numero_de_mortes.to_csv('paises-com-mais-mortes.csv')
```

```python
# Agrupando por País
paises_unicos = dados_covid['Country'].unique()

# Agrupando os dados
paises_agrupados = dados_covid.loc[:,['Lat','Long','Deaths']]

# Exportando os dados para gerar o mapa
paises_agrupados.to_csv('paises.csv')
```

Com Base nos dados acima podemos conluir que o País com o maior número de mortes foi o Estados Unidos da América tanto no período analisado que foi de 187 dias com os incrivéis número de  11.011.411 mortos.
